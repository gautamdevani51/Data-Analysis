## Company Sales Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Exercise 1: Read Total profit of all months and show it using a line plot
df = pd.read_csv("company_sales_data.csv")
x = df['month_number']
y = df['total_profit']
plt.figure(figsize=(10, 6))
plt.plot(x, y, color = 'blue', marker = '*', label = "Total_Profit")
plt.title("Total Profit of All Months")
plt.xlabel('Number of Months')
plt.ylabel("Total_Profit")
plt.xticks(x)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.legend(loc = 'upper left', fontsize = 12)
plt.tight_layout()
plt.grid()
plt.savefig('1_Total_Profit_of_All_Months.jpg')
plt.show()
```

```python
'''Exercise 2: Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: –

Line Style dotted and Line-color should be red
Show legend at the lower right location.
X label name = Month Number
Y label name = Sold units number
Add a circle marker.
Line marker color as read
Line width should be 3
'''
df = pd.read_csv("company_sales_data.csv")
x = df['month_number']
y = df['total_profit']
plt.figure(figsize=(10, 6))
plt.plot(x, y, color = 'red', markerfacecolor = "black", marker = 'o', linestyle = ':', 
         linewidth = 3, label = "Total_Profit")
plt.title("Total Profit of All Months")
plt.xlabel('Month Number')
plt.ylabel("Sold Units Number")
plt.xticks(x)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.legend(loc = 'lower right', fontsize = 12)
plt.tight_layout()
plt.grid()
plt.savefig('2_Total_Profit_All_Months_With_style.jpg')
plt.show()
```

```python
'''Exercise 3: Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. 
(i.e., Separate Plotline for each product ).
'''
df = pd.read_csv("company_sales_data.csv")
facecream = df['facecream'].to_list()
facewash = df['facewash'].to_list()
toothpaste = df['toothpaste'].to_list()
bathingsoap = df['bathingsoap'].to_list()
shampoo = df['shampoo'].to_list()
moisturizer = df['moisturizer'].to_list()
x = df['month_number']
plt.figure(figsize=(10, 6))
plt.plot(x, facecream, color = 'blue', marker = 'o', label = 'Face_Cream')
plt.plot(x, facewash, color = 'orange', marker = 'o', label = 'Face_Wash')
plt.plot(x, toothpaste, color = 'green', marker = 'o', label = 'Tooth_Paste')
plt.plot(x, bathingsoap, color = 'red', marker = 'o', label = 'Bathing_Soap')
plt.plot(x, shampoo, color = 'purple', marker = 'o', label = 'Shampoo')
plt.plot(x, moisturizer, color = 'brown', marker = 'o', label = 'moisturizer')
plt.xticks(x)
plt.xlabel("Month Number")
plt.ylabel("Sales of Each Product")
plt.title("Sales Data of All Products")
plt.legend(loc = "upper left", fontsize = 12)
plt.tight_layout()
plt.savefig('3_All_Product_Sales_Data.jpg')
plt.show()
```

```python
'''
Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.
'''
df = pd.read_csv("company_sales_data.csv")
x = df['month_number']
y = df['toothpaste']
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color = 'blue', label = 'Tooth_Paste')
plt.xticks(x)
plt.xlabel("Month Number")
plt.ylabel("Sales of Toothpaste")
plt.title("Sales Data of Toothpaste")
plt.legend(loc = "upper left", fontsize = 12)
plt.tight_layout()
plt.grid(True, linestyle = '-')
plt.savefig('4_Toothpaste_Sales_Data.jpg')
plt.show()
```

```python
'''
Exercise 5: Read face cream and facewash product sales data and show it 
using the bar chart
The bar chart should display the number of units sold per month for each product. 
Add a separate bar for each product in the same chart.
'''
df = pd.read_csv("company_sales_data.csv")
x = df['month_number']
plt.figure(figsize=(10, 6))
plt.bar(x, df['facecream'], color = 'red', label = "Facecream Data", align="edge")
plt.bar(x, df['facewash'], color = 'green', label = "Facewash Data", align="edge")
plt.xlabel("Month Number")
plt.ylabel("Sales Per Month")
plt.xticks(x)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title("Facecream And Facewash Sales Data")
plt.legend()
plt.tight_layout()
plt.savefig('5_Facecream_and_facewash_data.jpg')
plt.show()
```

```python
'''
Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. 
Save this plot to your hard disk
'''
df = pd.read_csv("company_sales_data.csv")
x = df['month_number']
plt.figure(figsize=(10, 6))
plt.bar(x, df['bathingsoap'], color = 'red', label = "Bathing Soap Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Sales of Bathing Soap")
plt.xticks(x)
plt.grid(True, linestyle="-")
plt.title("Bathing Soap Sales Data")
plt.legend()
plt.tight_layout()
plt.savefig('6_Bathing_Soap_Sales_data.jpg')
plt.show()
```

```python
'''
Exercise 7: Read the total profit of each month and show it using the histogram to see the 
most common profit ranges
'''
df = pd.read_csv("company_sales_data.csv")
plt.figure(figsize=(10, 6))
profitList = df ['total_profit'].tolist()
labels = ['low', 'average', 'Good', 'Best']
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitList, profit_range, label = 'Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Profit data')
plt.savefig('7_Total_Profit_of_Each_Month.jpg')
plt.show()
```

```python
'''
Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

'''
df = pd.read_csv("company_sales_data.csv")
monthList  = df ['month_number'].tolist()

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.figure(figsize=(10, 6))
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.savefig("8_Total_Sales_of_Each_Product.jpg")
plt.show()
```

```python
# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot
df = pd.read_csv("company_sales_data.csv")

monthList  = df ['month_number'].tolist()
bathingsoap   = df ['bathingsoap'].tolist()
faceWashSalesData   = df ['facewash'].tolist()

f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(monthList, bathingsoap, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(monthList, faceWashSalesData, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')
plt.figure(figsize=(10, 6))
plt.xticks(monthList)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.savefig("9_Subplot_of_Bathingsoap_And_Facewash.jpg")
plt.show()
```

```python
# Exercise Question 10: Read all product sales data and show it using the stack plot

df = pd.read_csv("company_sales_data.csv")
plt.figure(figsize=(10, 6))
monthList  = df ['month_number'].tolist()
faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingsoapSalesData   = df ['bathingsoap'].tolist()
shampooSalesData   = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

plt.stackplot(monthList, faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.savefig("10_SalesData_For_All_products.jpg")
plt.show()
```
